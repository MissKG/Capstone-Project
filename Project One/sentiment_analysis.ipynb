```python
# Importing neccessary packages
import pandas as pd
import numpy as np
import spacy

# Loding spaCy language model
nlp = spacy.load('en_core_web_md')
```

```python
# Loading the dataset
amazon = pd.read_csv('amazon_product_reviews.csv')
amazon.head()
```

```python
# Selecting our feature coloumn
reviews_data = amazon[['reviews.text']]
reviews_data
```

```python
# Checking for any null values
reviews_data.isnull().sum()
```

```python
def preprocess(text):
    """
    This is a function for preprocessing the text data; tokenisation, 
    lemmatisation, and standardising the text data
    """
    doc = nlp(text.lower().strip())
    processed = []
    for token in doc:
        if not token.is_stop and not token.is_punct:
            processed.append(token.lemma_.lower()) # Adding the processed 
                                                   # text data to the 'processed' list

    return ' '.join(processed) # Rejoining the tokens into strings
```

```python
# Using the 'apply' function to apply the preprocessing to our 
# feature column, to form a new column
reviews_data['processed.text']= reviews_data['reviews.text'].apply(preprocess)
```

```python
# Checking our preprocessing has been applied well
reviews_data.head()
```

```python
# Importing neccessary package for sentiment 
# analysis
from textblob import TextBlob
```

```python
# Accessing our text data from the new feature column
data = reviews_data['processed.text'].values
data
```

```python
def analyse_polarity(text):
    """
    This is a function for analysing the polarity of 
    our text data. 
    """
    # Using 'TextBlob' find the polarity of our
    # text data
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    # Our function returns the polarity of the
    # text data
    return polarity
```

```python
# An empty list to contain the assigned sentiments
sentiments = []

# Iterating through our text data in the list of reviews
for item in data:
    # Calling our 'analyse_polarity' function and passing 
    # each review in the list as our arguement
    polarity_score = analyse_polarity(item) 

    # Conditions for sentiment assignment
    if polarity_score > 0:
        sentiment = "positive"
    elif polarity_score < 0:
        sentiment = "negative"
    else:
        sentiment = "neutral"

    # Adding each sentiment assignment to our 'sentiments' list
    sentiments.append(sentiment)
```

```python
# Viewing our 'sentiments' list
sentiments
```

```python
# Accessing the 3 different sentiments in our sentiments list 
# and assigning them to their relevant variables
positive_count = sentiments.count("positive")
negative_count = sentiments.count("negative")
neutral_count = sentiments.count("neutral")

# Getting the total number of sentiments in the
# sentiments list
total = len(sentiments)

# Getting the percentage of each sentiment in the 
# sentiments list, assigning to relevant variables
positive_perc = (positive_count / total) * 100
negative_perc = (negative_count / total) * 100
neutral_perc = (neutral_count / total) * 100

# Printing the calculated sentiment percentages
print(f"Positive percentage of sentiments: {positive_perc:.2f}%")
print(f"Negative percentage of sentiments: {negative_perc:.2f}%")
print(f"Neutral percentage of sentiments: {neutral_perc:.2f}%")
```

```python
def predict_sentiment(sample_review):
    """
    This is a function for testing sample reviews on our model, 
    calling on our preprocessing and polarity analysis functions
    """
    preprocessed_review = preprocess(sample_review)
    polarity_score = analyse_polarity(preprocessed_review)
    
    # Conditions for sentiment assignment
    if polarity_score > 0:
        sentiment = "Positive"
    elif polarity_score < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    # Returning the assigned sentiment
    return sentiment
```

```python
# Getting user input for sample review testing
test_review = input("Enter your sample review: ")

# Calling on our sample review testing function and 
# passing the review from our user input as an arguement
test_predicted_sentiment = predict_sentiment(test_review)
print("Predicted sentiment for your sample review is:", test_predicted_sentiment)
```
